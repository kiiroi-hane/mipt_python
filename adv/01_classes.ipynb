# Функциональное

Хотя питон изначально не задумывался как язык для функционального программирования, в нём **есть достаточно большой арсенал**, позволяющий разработку функционально-ориентированных программ.

````python
lambda_function = lambda x, y : 2*(x + y)
seq = (1, 2, 3, 4, 5, 6, 7, 8, 9)
seq2 = (5, 6, 7, 8, 9, 0, 3, 2, 1)
result = map(lambda_function, seq, seq2)
result
````

Помимо дефолтных лямбда функций и map есть **filter**:

`filter(предикативная_функция, последовательность)`

Предикативная функция - это функция, по которой мы фильтруем последовательность; на входе у неё элемент последовательности, на выходе - 1, если элемент лежит в последовательности, 0 - если нет

````python
is_even = lambda x: x % 2 == 0
seq = (1, 2, 3, 4, 5, 6, 7, 8, 9)
filtered = filter(is_even, seq)
list(filtered)
````

# Задача

Нужно вывести отсортированный массив. Какое из решений  - функциональный подход, а какое - процедурный?

````python
# решение 1

INPUT = [1, 2, 5, 12, 3, 5, 2, 7, 12]


def main():
    numbers = INPUT[:]
    filter_and_sort(numbers)  # Сортируем и фильтруем по месту
    for number in numbers:
        print(number)         # Выводим поэлементно в цикле


def filter_and_sort(values):
    values.sort()  # Список сортируется по месту
    previous_value = None
    index = 0
    while index < len(values):
        value = values[index]
        if value == previous_value and index > 0:
            # Элемент удаляется из списка, то есть список опять модифицируется
            values.pop(index)
        else:
            index += 1
            previous_value = value
````

# Наследование

Пусть нам нужно создать классы помидоров, огурцов и баклажанов. Заметим, что код у этих классов очень похожий (так  как и суть похожа - поливать нужно всех, у всех нужно знать количество растений), и происходит много копирований и вставки.

````python
class Tomato():
    current_water_amount = 0
    veggies_number = 0

    def __init__(self):
        self.current_water_amount = 0
        self.veggies_number = 1

    def water(self):
        self.current_water_amount += 8

    def take_veggies(self):
        self.veggies_number = 0

    def add_veggie(self):
        self.veggies_number += 1


class Cucumber():
    current_water_amount = 0
    veggies_number = 0

    def __init__(self):
        self.current_water_amount = 0
        self.veggies_number = 1

    def water(self):
        self.current_water_amount += 8

    def take_veggies(self):
        self.veggies_number = 0

    def add_veggie(self):
        self.veggies_number += 1


class Eggplant():
    current_water_amount = 0
    veggies_number = 0

    def __init__(self):
        self.current_water_amount = 0
        self.veggies_number = 1

    def water(self):
        self.current_water_amount += 8

    def take_veggies(self):
        self.veggies_number = 0

    def add_veggie(self):
        self.veggies_number += 1
````

Чтобы избежать этого, мы можем ввести родительский класс (потенциально - абстрактный класс). В нём определяется все общее для этих классов и копипасты не происходит.

````python
class Plant():
  current_water_amount = 0
  veggies_number = 0

  def __init__(self):
    self.current_water_amount = 0
    self.veggies_number = 1

  def take_veggies(self):
    self.veggies_number = 0

  def add_veggie(self):
    self.veggies_number += 1


class Tomato(Plant):
    def water(self):
      self.current_water_amount += 8

class Cucumber(Plant):
    def water(self):
      self.current_water_amount += 10

class Eggplant(Plant):
    def water(self):
      self.current_water_amount += 3
````

````python
cucumber = Cucumber()
cucumber.water()
cucumber.current_water_amount
````

# Инкапсуляция

Представим, что у нас разработана библиотека palisadnik с нашими огурцами, томатами и прочими растениями. Теперь мы отдаём её в прод. Но не будем ли мы бояться, что когда люди будут ей пользоваться, они могут случайно что-то испортить и сломать работу у себя?

Для упрощения работы, методы можно "прятать" от окружающих. Давайте посмотрим, как это делается.

````python
class Plant():
    variety_name = ''
    __days_to_ripen = 30

    def __init__(self, name):
        self.variety_name = name

    def get_variety(self):
        print('Сорт этого растения - ' + self.variety_name)

    def _prune(self):
        print('Вы зашли в метод prune - формирование растения. Делайте это с умом и осторожностью')

    def __ripen(self):
        print('Нельзя заставить овощ созреть раньше времени!')
````

````python
plant = Plant('Бычье сердце')

plant.get_variety()
plant._prune()
````

# Антипаттерны

**Антипаттерн** — это распространённый, но неэффективный или контрпродуктивный подход к решению типичных проблем в программировании.

В ООП в питоне есть несколько примеров, как МОЖНО делать, но НЕ НУЖНО.

## 1. Не стоит менять атрибуты класса/объекта после конструирования (init)

````python
# ПЛОХО

class A:
    pass

a = A()
a.x = 25
print(a.x)
````

````python
# ХОРОШО

class A:
    def __init__(self, x):
        self.x = x

a = A(25)
print(a.x)
````

## 2. Не стоит обращаться к атрибутам объекта, не объявленным в init

````python
# ПЛОХО

class A:
    x: int = 0

    def foo(self):
        self.x += 1

a = A()
b = A()
a.foo()
a.foo()
a.x, b.x, A.x
````

````python
class A:
    def __init__(self):
        self.x = 0

    def foo(self) -> None:
        self.x += 1

a = A()
a.foo()
a.foo()
a.x
````

## 3. Не стоит делать атрибуты класса (не объекта) мутабельными объектами

````python
# ПЛОХО

class A:
    names: list[str] = []

    def add_name(self, name: str) -> None:
        self.names.append(name)

a = A(); b = A()

a.add_name('Ilariia')
b.add_name('Lesha')
a.names, b.names
````

````python
# ХОРОШО

class A:
    def __init__(self):
        self.names: list[str] = []

    def add_name(self, name: str) -> None:
        self.names.append(name)

a = A(); b = A()

a.add_name('Ilariia')
b.add_name('Lesha')
a.names, b.names
````

Иногда хочется, чтобы у объекта класса был атрибут, который принимает значение из некоторого списка возможных значений, логично обоснованных.

Например, для томатов это его статус жизни - он ещё саженец, большое растение или уже умер.

Как решить такую задачку?

````python
SEED = 0
SPROUT = 1
SEEDLING = 2
MATURE = 3
FLOWERING = 4
FRUITING = 5
HARVESTED = 6
DISEASED = 7
DEAD = 8


class Tomato:
    def __init__(self, variety: str):
        self.variety = variety
        self.status = 0  # SEED
        self.days = 0

    def grow(self):
        """Увеличивает возраст и меняет статус растения"""
        self.days += 1

        # Магические числа в условиях
        if self.status == 0 and self.days > 5:  # SEED -> SPROUT
            self.status = 1
        elif self.status == 1 and self.days > 10:  # SPROUT -> SEEDLING
            self.status = 2
        elif self.status == 2 and self.days > 25:  # SEEDLING -> MATURE
            self.status = 3
        elif self.status == 3 and self.days > 40:  # MATURE -> FLOWERING
            self.status = 4
        elif self.status == 4 and self.days > 60:  # FLOWERING -> FRUITING
            self.status = 5

    def get_status_name(self) -> str:
        """Возвращает читаемое название статуса"""
        status_names = {
            0: "Семя",
            1: "Росток",
            2: "Рассада",
            3: "Взрослое растение",
            4: "Цветение",
            5: "Плодоношение",
            6: "Урожай собран",
            7: "Болеет",
            8: "Мёртвое"
        }
        return status_names.get(self.status, "Неизвестный статус")

    def can_harvest(self) -> bool:
        """Можно ли собирать урожай?"""
        return self.status == 5  # FRUITING

    def harvest(self):
        """Собрать урожай"""
        if self.status != 5:  # FRUITING
            raise ValueError("Нельзя собрать урожай - нет плодов")
        self.status = 6  # HARVESTED
````

В чём минус этого подхода?

Нечитаемость статусов, которые надо постоянно помнить. Легко ошибиться в номере статуса - и потом это придётся дебажить :(

````python
from enum import Enum, auto


class TomatoStatus(Enum):
    SEED = auto()
    SPROUT = auto()
    SEEDLING = auto()
    MATURE = auto()
    FLOWERING = auto()
    FRUITING = auto()
    HARVESTED = auto()
    DISEASED = auto()
    DEAD = auto()

    def display_name(self) -> str:
        """Возвращает читаемое название статуса"""
        names = {
            TomatoStatus.SEED: "Семя",
            TomatoStatus.SPROUT: "Росток",
            TomatoStatus.SEEDLING: "Рассада",
            TomatoStatus.MATURE: "Взрослое растение",
            TomatoStatus.FLOWERING: "Цветение",
            TomatoStatus.FRUITING: "Плодоношение",
            TomatoStatus.HARVESTED: "Урожай собран",
            TomatoStatus.DISEASED: "Болеет",
            TomatoStatus.DEAD: "Мёртвое"
        }
        return names[self]


class Tomato:
    def __init__(self, variety: str):
        self.variety = variety
        self.status = TomatoStatus.SEED
        self.days = 0
        self.water_count = 0

    def grow(self):
        """Увеличивает возраст и меняет статус растения по правилам"""
        if self.status in [TomatoStatus.DEAD, TomatoStatus.HARVESTED]:
            return  # Мёртвое или собранное растение не растёт

        self.days += 1
        next_status = self._determine_next_status()

        if next_status and self.status.can_grow_to(next_status):
            old_status = self.status
            self.status = next_status
            print(f"{self.variety}: {old_status.display_name()} → {next_status.display_name()}")

    def _determine_next_status(self) -> TomatoStatus:
        """Определяет следующий статус на основе возраста и условий"""
        growth_stages = {
            5: TomatoStatus.SPROUT,      # Через 5 дней из семени в росток
            10: TomatoStatus.SEEDLING,   # Через 10 дней в рассаду
            25: TomatoStatus.MATURE,     # Через 25 дней во взрослое растение
            40: TomatoStatus.FLOWERING,  # Через 40 дней в цветение
            60: TomatoStatus.FRUITING    # Через 60 дней в плодоношение
        }

        # Проверяем условия для перехода на следующую стадию
        for days_required, next_stage in growth_stages.items():
            if self.days >= days_required and self.status.value < next_stage.value:
                return next_stage

        return None
````

Альтернативно можно явно указать соответствие, не используя auto()

````python
from enum import Enum

class TomatoStatus(Enum):
    SEED = 0
    SPROUT = 1
    SEEDLING = 2
    MATURE = 3
    FLOWERING = 4
    FRUITING = 5
    HARVESTED = 6
    DISEASED = 7
    DEAD = 8
````

Иногда это может вызывать ошибки в коде. Для этого можно использовать декоратор unique из пакета Enum

````python
from enum import Enum, unique

@unique
class TomatoStatus(Enum):
    SEED = 1
    SEED_TWO = 1
````
